# Main Analysis

```python
n_outcomes = 2000 # number of bootstrapped portfolios per regression datapoint
pf_sizes = np.arange(2, 32 + 1, 3) # portfolio sizes 
pf_horizons = np.arange(1, 25 + 1, 3) # investment horizons

# initialize bootstrap object
B = cc.BootstrapClass(
    return_data_path=f'./DATA/ret_full.csv', 
    size_data_path=f'./DATA/sz_full.csv', 
    sector=0, 
    pf_sizes=pf_sizes, 
    pf_horizons=pf_horizons, 
    n_outcomes=n_outcomes
)

# set data path
data_path = f"./REG_DATA/reg_data_sector0_{n_outcomes}.csv"

# if a file with regression data exists already, then use that instead of running the bootstrap simulation
if Path(data_path).is_file():
    B.run_regression(data_path=data_path)
else:
    B.run_regression()
```

# Robustness (Sector) Tests

```python
sectors = ['25', '20', '15', '45', '35', '40', '55', '10', '50', '30', '60']

# check if there are enough stocks in each sector
for sector in sectors:

    # initialize bootstrap object
    b = cc.BootstrapClass(
        return_data_path=f'./DATA/ret_sector{int(sector)}.csv', 
        size_data_path=f'./DATA/sz_sector{int(sector)}.csv',
        sector=sector
    )

    assert b.min_valid > np.max(pf_sizes) 

    print(f"Sector {int(sector)}, min_valid = {b.min_valid}")
```

```python

for sector in sectors:
    print(f"\nRunning for sector {int(sector)}...")

    # initialize bootstrap object
    b = cc.BootstrapClass(
        return_data_path=f'./DATA/ret_sector{int(sector)}.csv', 
        size_data_path=f'./DATA/sz_sector{int(sector)}.csv', 
        sector=sector, 
        pf_sizes=pf_sizes, 
        pf_horizons=pf_horizons,  
        n_outcomes=n_outcomes
    )

    # set data path
    data_path = f"./REG_DATA/reg_data_sector{int(sector)}_{n_outcomes}.csv"

    # if a file with regression data exists already, then use that instead of running the bootstrap simulation
    if Path(data_path).is_file():
        b.run_regression(data_path=data_path)
    else:
        b.run_regression()
```
