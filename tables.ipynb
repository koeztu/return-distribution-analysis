# Robustness Test Tables

Simply read the regression tables and gather all the relevant values.

```python
# SKEW

n_outcomes = 2000
sectors = [0, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60]
sector_labels = {0: "All", 10: "10", 15: "15", 20: "20", 25: "25", 30: "30",
                 35: "35", 40: "40", 45: "45", 50: "50", 55: "55", 60: "60"}
variables = ["constant", "value", "minvar", "riskpar", "rand"]
col_names = ["Sector", "Constant", "Val", "Min Var", "Risk Par", "Rand"]

def significance_stars(p):
    if p < 0.001:
        return "***"
    elif p < 0.01:
        return "**"
    elif p < 0.05:
        return "*"
    elif p < 0.1:
        return "."
    return ""

def format_coeff(coeff, p):
    return f"{coeff:.5g}{significance_stars(p)}"

rows = []
raw = {}  # store raw coefficients for High/Low/Diff

for sector in sectors:
    data_path = f"./REG_RESULTS/results_sector{int(sector)}_skew.csv"
    df = pd.read_csv(data_path, index_col=0)

    row = [sector_labels[sector]]
    sector_raw = {}
    for var in variables:
        coeff = df.loc[var, "Coefficient"]
        p = df.loc[var, "p-value"]
        row.append(format_coeff(coeff, p))
        sector_raw[var] = coeff
    rows.append(row)
    raw[sector] = sector_raw

# compute High, Low, Diff across sectors (excluding "All" which is sector 0)
sector_only = [s for s in sectors if s != 0]
high_row = ["High"]
low_row = ["Low"]
diff_row = ["Diff"]

for var in variables:
    vals = {s: raw[s][var] for s in sector_only}
    high_sector = max(vals, key=vals.get)
    low_sector = min(vals, key=vals.get)
    high_row.append(str(high_sector))
    low_row.append(str(low_sector))
    diff_row.append(f"{abs(vals[high_sector] - vals[low_sector]):.5g}")

rows.append(high_row)
rows.append(low_row)
rows.append(diff_row)

out_df = pd.DataFrame(rows, columns=col_names)
out_df.to_csv("./TABLES/method_coefficients_by_sector_skew.csv", index=False)
```
